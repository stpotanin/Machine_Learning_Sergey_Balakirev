```python
import numpy as np
import matplotlib.pyplot as plt

# Определяем функцию и её производную
f = lambda x: 0.4 * x + 0.1 * np.sin(2 * x) + 0.2 * np.cos(3 * x) # нужна только для написания df()
df = lambda x: 0.4 + 0.2 * np.cos(2 * x) - 0.6 * np.sin(3 * x)

η = 1.0
N = 100  # число итераций градиентного алгоритма
γ = 0.7

# Запускаем три метода в одном цикле
x = 4.0
x_Nesterov = []
x_imp = 4.0
x_impulse = []
x_wi = 4.0
x_without_impulse = []

# Инициализация импульса
v_nesterov = 0   # начальное значение для Nesterov
v_impulse = 0    # начальное значение для импульса

for _ in range(N):
    # Nesterov
    v_nesterov = γ * v_nesterov + (1 - γ) * η * df(x - γ * v_nesterov)
    x -= v_nesterov
    x_Nesterov.append(x)

    # Impulse
    v_impulse = γ * v_impulse + (1 - γ) * η * df(x_imp)
    x_imp -= v_impulse
    x_impulse.append(x_imp)

    # Without impulse
    x_wi -= η * df(x_wi)
    x_without_impulse.append(x_wi)

# График сходимости
plt.plot(x_Nesterov, label='Nesterov')
plt.plot(x_impulse, label='Impulse')
plt.plot(x_without_impulse, label='Without impulse')
plt.grid()
plt.title('График сходимости')
plt.xlabel('Номер итерации')
plt.ylabel('Значение x')
plt.legend()
plt.show()
```
