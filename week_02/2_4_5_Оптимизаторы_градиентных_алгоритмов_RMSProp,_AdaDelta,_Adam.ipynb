```python
import numpy as np


df = lambda x: -.5 + .4 * x - .03 * x**2 - 1.2 * np.cos(4 * x)

η = 0.3
N = 200 # число итераций градиентного алгоритма
γ = 0.8
v = 0   # начальное значение для v

# Запускаем два спуска в однеом цикле
x = -3.5
x_impulse = []
x_wi = -3.5
x_without_impulse = []
for _ in range(N):
    # с импульсом
    v = γ * v + (1 - γ) * η * df(x)
    x -= v
    x_impulse.append(x)
    # без импульса
    x_wi -= η * df(x_wi)
    x_without_impulse.append(x_wi)

# График сходимости
import matplotlib.pyplot as plt
plt.plot(x_impulse)
plt.plot(x_without_impulse)
plt.show()
```
