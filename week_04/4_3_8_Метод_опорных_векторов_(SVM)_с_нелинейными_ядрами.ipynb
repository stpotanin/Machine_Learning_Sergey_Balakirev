```python
import numpy as np
from sklearn import svm

func = lambda x: np.sin(0.5*x) + 0.2 * np.cos(2*x) - 0.1 * np.sin(4 * x) + 3

coord_x = np.expand_dims(np.arange(-4.0, 6.0, 0.1), axis=1) #  добавили новое измерение к массиву
coord_y = func(coord_x).ravel() # многомерный массив в непрерывный плоский одномерный массив
x_train = coord_x[::3]          # каждый третий
y_train = coord_y[::3]

svr = svm.SVR(kernel='rbf').fit(x_train, y_train)    # SVM с нелинейным ядром + обучение

predict = svr.predict(coord_x)

Q = ((predict - coord_y)**2).mean()

# График
import matplotlib.pyplot as plt

plt.plot(coord_x, coord_y, label='Функция')
plt.plot(coord_x, predict, 'k--', label='Модель', color='red')
plt.legend()
plt.show()
```
