```python
# @title SVR (Support Vector Regression)
import numpy as np
from sklearn import svm

def func(x):
    return np.sin(0.5*x) + 0.2 * np.cos(2*x) - 0.1 * np.sin(4 * x) - 2.5


def model(w, x):
    return w[0] + w[1] * x + w[2] * x ** 2 + w[3] * x ** 3 + w[4] * np.cos(x) + w[5] * np.sin(x)


# обучающая выборка
coord_x = np.arange(-4.0, 6.0, 0.1)
coord_y = func(coord_x)

x_train = np.array([[x, x**2, x**3, np.cos(x), np.sin(x)] for x in coord_x])
y_train = coord_y

svr = svm.SVR(kernel='linear')
svr.fit(x_train, y_train)
w = [svr.intercept_[0]] + list(svr.coef_[0])
Q = np.mean([(model(w, x) - y)**2 for x, y in zip(coord_x, coord_y)])

# График
import matplotlib.pyplot as plt

plt.scatter(coord_x, coord_y, label='Данные')
plt.plot(coord_x, model(w, coord_x), color='red', label='Модель')
plt.xlabel('x')
plt.ylabel('func(x)')
plt.title('SVR (Support Vector Regression)')
plt.legend()
plt.show()
```
