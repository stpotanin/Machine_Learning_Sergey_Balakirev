> "Математика не терпит вульгарности и в ней должно быть всё чётко определено." © Сергей Балакирев

поэтому:

> Градиент — это вектор, который характеризует величину и направление наибольшего роста функции. Он определяется как набор частных производных функции по всем её переменным и указывает направление наиболее быстрого изменения функции в данной точке.

В нашем примере признак один, одномерное пространство (ось х). Поэтому градиент будет совпадать со своей единственной компонентой. Геометричекси это вектор, параллельный единственной оси призаков - х.





```python
import time
import numpy as np
import matplotlib.pyplot as plt
from IPython import display

# f = lambda x: x*x - 5*x + 5
# df = lambda x: 2*x - 5
f = lambda x: np.sin(x) + .5*x
df = lambda x: np.cos(x) + .5

N = 20
xx = 0
lmd = 0.3

# x_plt = np.arange(0, 5.0, 0.1)
x_plt = np.arange(-5, 5.0, 0.1)
f_plt = [f(x) for x in x_plt]

for i in range(N + 1):
    xx = xx - lmd * df(xx)

    fig, ax = plt.subplots(figsize=(5, 5))
    ax.grid(True)
    ax.plot(x_plt, f_plt, label='f(x)')
    ax.set_title(" ".join(["Gradient Descent", f"step №{i}"]))

    if i == N:
        ax.scatter(xx, f(xx), c='blue', label='Converged Point')
    else:
        ax.scatter(xx, f(xx), c='red', label='Current Point')

    ax.text(xx+0.5, f(xx)+0.5, f'xx = {xx:.2f}', fontsize=10, ha='right')

    ax.legend()
    plt.show()

    time.sleep(0.2)
    display.clear_output(wait=True)
```
