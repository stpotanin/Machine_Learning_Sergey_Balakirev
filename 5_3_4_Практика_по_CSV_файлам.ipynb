# Поиск по названию колонки

```python
import csv

def print_column_values(filename, colname):
    with open(filename, encoding='utf-8') as f:
        res = list(csv.reader(f))
        if colname in res[0]:
            indx = [i for i in range(len(res[0])) if res[0][i] == colname][0]
            [print(x[indx]) for x in res[1:]]
        else: print(f'Колонка "{colname}" не найдена.')
```

```python
# @title Короткая проверка
!wget -nc -q 'https://raw.githubusercontent.com/artemegorof/modules_and_files/refs/heads/main/module_05/lesson_03/step_04/library_books.csv'

print_column_values('library_books.csv', "Автор")

# вызов должен распечатать на экран следующее:

# Джордж Оруэлл
# Михаил Булгаков
# Эрих Мария Ремарк
# Фёдор Достоевский

print_column_values('library_books.csv', "автор")

# Колонка "автор" не найдена.
```
