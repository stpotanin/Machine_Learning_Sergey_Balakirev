```python
import numpy as np

func = lambda x: .1 * x**2 - np.sin(x) + 0.1 * np.cos(x * 5) + 1

coord_x = np.arange(-5.0, 5.0, 0.1) # значения отсчетов по оси абсцисс
coord_y = func(coord_x) # значения функции по оси ординат
w = [1.11, -0.26, 0.061, 0.0226, 0.00178] # веса нам даны

# Добавляем признаки согласно модели a(x,w)=w_0+w_1⋅x+w_2⋅x^2+w_3⋅x^3+w_4⋅x^4
X = [(1, x, x**2, x**3, x**4) for x in coord_x] # единица впереди
# Реализуем модель a(x,w)=w_0+w_1⋅x+w_2⋅x^2+w_3⋅x^3+w_4⋅x^4
a = lambda x, w: sum(w_j * x_j for x_j, w_j in zip(x, w))
# Реализуем Q(a,X) с функцией потерь L_i(w,x_i) = (a(x_i,w)−f(x_i))^2
Q = np.mean([(a(x_i, w) - y)**2 for x_i, y in zip(X, coord_y)])
print(Q)
```

```python
import numpy as np

def func(x):
    return 0.1 * x**2 - np.sin(x) + 0.1 * np.cos(x * 5) + 1

coord_x = np.arange(-5.0, 5.0, 0.1) # значения отсчетов по оси абсцисс
coord_y = func(coord_x)             # значения функции по оси ординат
w = [1.11, -0.26, 0.061, 0.0226, 0.00178] # веса нам даны

X = np.vander(coord_x, len(w), increasing=True) # добавляем признаки по модели a
Q = ((X @ w - coord_y)**2).mean()
Q
```

```python
# Vadim Kopeykin https://stepik.org/lesson/1370055/step/6?discussion=10495300&thread=solutions&unit=1386245
import numpy as np

func = lambda x: 0.1 * x**2 - np.sin(x) + 0.1 * np.cos(x * 5) + 1.

x = np.arange(-5.0, 5.0, 0.1) # значения отсчетов по оси абсцисс
y = func(x)                   # значения функции по оси ординат

w = np.array([1.11, -0.26, 0.061, 0.0226, 0.00178]) # параметры модели
X = x.reshape(-1, 1) ** np.arange(len(w))           # матрица признаков

Q = np.mean((X @ w - y)**2) # показатель качества работы модели
Q
```
